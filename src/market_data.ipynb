```python
# 'yfinance' is a third-party Python library that provides a convenient wrapper
# around Yahoo Finance's publicly available market data.
import yfinance as yf

# Ticker is a class provided by the yfinance library.
# It represents a single financial instrument identified by its ticker symbol.
ticker = yf.Ticker("GOOGL")

# Fetch historical market data for the specified time range.
# period="5y" requests approximately five years of historical data.

# interval="1d" specifies daily price data for each trading day.
df = ticker.history(period="5y", interval="1d")

# Display the resulting DataFrame, which typically includes:
# Open, High, Low, Close, Volume, Dividends, and Stock Splits
df
```

```python
def validate_price_data(df):
    """
    Validate and sanitize raw price data returned from a market data provider.

    Ensures:
    - Data is not empty
    - Required OHLCV columns are present
    - Rows with missing values are removed

    Returns a cleaned DataFrame safe for downstream calculations.
    """
    if df.empty:
        # Raise an exception if the provider returned no data.
        raise ValueError("No data returned.")

    # Python sets are unordered data structures, and do not store duplicate values.
    required_cols = {"Open", "High", "Low", "Close", "Volume"}

    # 'df' holds a reference to a pandas DataFrame.
    # The .columns property will return an Index object which contains all the
    # column names in that DataFrame.

    # .issubset() this is a set method.
    # Returns True if all elements in required_cols are in df.columns.
    # Otherwise, it returns False.

    # The 'not' operator reverses the logical truth of an operand.

    # Ensure all required columns are present before proceeding.
    if not required_cols.issubset(df.columns):
        raise ValueError("Missing required columns")

    # Drop rows in the DataFrame where any columns contains 'NaN'.
    # to prevent incorrect calculations.
    df = df.dropna()

    return df
```
